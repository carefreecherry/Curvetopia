```python
# Example detections from YOLOv5 output
detections = [
    {'image': 'Screenshot 2024-08-11 164932.png', 'shapes': []},
    {'image': 'Screenshot 2024-08-11 165055.png', 'shapes': []},
    {'image': 'Screenshot 2024-08-11 165113.png', 'shapes': ['circle', 'ellipse']},
    {'image': 'Screenshot 2024-08-11 165142.png', 'shapes': ['square', 'star']},
    {'image': 'Screenshot 2024-08-11 165213.png', 'shapes': []},
    {'image': 'Screenshot 2024-08-11 165235.png', 'shapes': ['circle']},
    {'image': 'WhatsApp Image 2024-08-11 at 05.36.25_254dc031.jpg', 'shapes': ['square', 'star']},
    {'image': 'WhatsApp Image 2024-08-11 at 05.37.44_039b17f5.jpg', 'shapes': ['circle']},
    {'image': 'WhatsApp Image 2024-08-11 at 05.39.01_ef3ada49.jpg', 'shapes': ['circle', 'ellipse']},
    {'image': 'WhatsApp Image 2024-08-11 at 05.39.29_ad16a8fa.jpg', 'shapes': []},
]

# Function to return the symmetry details for a given shape
def get_symmetry_info(shape):
    symmetry_dict = {
        'circle': 'infinite lines of symmetry',
        'rectangle': '2 lines of symmetry',
        'square': '4 lines of symmetry',
        'ellipse': '2 lines of symmetry',  # Major and minor axis
        'star': '5 lines of symmetry',  # assuming a typical 5-pointed star
        'hexagon': '6 lines of symmetry',
    }
    return symmetry_dict.get(shape, 'unknown symmetry')

# Function to print the symmetry details for each image
def print_symmetry_info(detections):
    for detection in detections:
        image_name = detection['image']
        shapes = detection['shapes']
        if shapes:
            for shape in shapes:
                symmetry_info = get_symmetry_info(shape)
                print(f"Image '{image_name}' has 1 {shape}, it has {symmetry_info}.")
        else:
            print(f"Image '{image_name}' has no detectable shapes.")

# Displaying results
print_symmetry_info(detections)
```
